# **FoxNet v1.0**

**Instructions**

1. Upload a fox image with the 'Upload' button.
2. Classify image with the 'Classify' button.
3. Repeat until fully satisfied.

```
path = Path()
learn_inf = load_learner(path/'FoxNet_export.pkl', cpu=True)
btn_upload = widgets.FileUpload()
out_pl = widgets.Output()
lbl_pred = widgets.Label()
btn_run = widgets.Button(description='Classify')
```

```
def on_click_classify(change):
    img = PILImage.create(btn_upload.data[-1])
    out_pl.clear_output()
    with out_pl: display(img.to_thumb(128,128))
    pred,pred_idx,probs = learn_inf.predict(img)
    # Re-label predictions
    if pred == 'arctic fox':
      pred = 'Arctic Fox (Vulpes lagopus)'
    if pred == 'cape fox':
      pred = 'Cape Fox (Vulpes chama)'
    if pred == 'corsac fox':
      pred = 'Corsac Fox (Vulpes corsac)'
    if pred == 'red fox':
      pred = 'Red Fox (Vulpes vulpes)'
    
    confidence_level = round((float(probs[pred_idx]) * 100), 2) # Set probability figure to value out of 100
    lbl_pred.value = f'Prediction: {pred}; Confidence level: {confidence_level}%'
    btn_upload.value.clear()
    btn_upload._counter = 0

btn_run.on_click(on_click_classify)
```

```
display(VBox([widgets.Label('Upload your fox image!'), 
      btn_upload, btn_run, out_pl, lbl_pred]))
```

# **Introducing FoxNet**
**The Amazing Fox Classifier**

It takes a cunning photographer to catch the likeness of the cunning fox, but it takes even more cunning to recognize the species... 

Do you have a photo of an adorable fox, and have a burning desire to know the species. Do you find yourself lost for words while trying to describe the species of fox that you saw stealing your chickens to your friends/pest control?

For those of us who don't spend our spare time practicing our fox species recognition skills, FoxNet offers fast and reliable fox classification - anytime, anywhere.*

Simply upload an image of a fox, and FoxNet will identify if the species is one of the following:
1. Red Fox (Vulpes vulpes)
2. Arctic Fox (Vulpes lagopus)
3. Corsac Fox (Vulpes corsac)
4. Cape Fox (Vulpes chama)

Note: FoxNet works best with colour photos taken of foxes from a relatively close distance.

Notes^2: FoxNet is only able to accurately classify foxes belonging to one of the four species listed above.

*(with an internet connection)
